```python
import pandas as pd

# Load the dataset
df_combined = pd.read_csv("01_data/raw/combined_dataset.csv")

# Display first few rows
print("Combined Dataset Head:")
print(df_combined.head())

# Check label distribution
print("\nLabel Distribution:")
print(df_combined['label'].value_counts())

# Check unique values in label column
print("\nUnique label values:")
print(df_combined['label'].unique())
```
