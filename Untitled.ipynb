```python
import pandas as pd

# Check combined dataset
df = pd.read_csv("01_data/raw/combined_dataset.csv")
print("Combined Dataset:")
print(f"Total samples: {len(df)}")
print(f"Label distribution:\n{df['label'].value_counts()}")
print(f"Unique labels: {df['label'].unique()}")

# Check train split
train_df = pd.read_csv("01_data/splits/train.csv")
print("\nTrain Split:")
print(f"Total: {len(train_df)}")
print(f"Label distribution:\n{train_df['label'].value_counts()}")

# Check val split
val_df = pd.read_csv("01_data/splits/val.csv")
print("\nValidation Split:")
print(f"Total: {len(val_df)}")
print(f"Label distribution:\n{val_df['label'].value_counts()}")

# Check test split
test_df = pd.read_csv("01_data/splits/test.csv")
print("\nTest Split:")
print(f"Total: {len(test_df)}")
print(f"Label distribution:\n{test_df['label'].value_counts()}")
```
