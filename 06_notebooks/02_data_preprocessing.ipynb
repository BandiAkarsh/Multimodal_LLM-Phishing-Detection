# Data Preprocessing: Extract Multimodal Features
This notebook extracts screenshots, HTML, and URL features from the dataset.

```python
import sys
import os

# Add the utils directory to the path
sys.path.append(os.path.abspath('../05_utils'))

import yaml
import asyncio
from data_preparation import DataPreprocessor
```

```python
# Load configuration
with open('../07_configs/config.yaml', 'r') as f:
    config = yaml.safe_load(f)

print("Configuration loaded successfully")
```

```python
# Initialize preprocessor
preprocessor = DataPreprocessor(config)
print("Preprocessor initialized")
```

```python
# Process training data (start with a small sample)
train_results = await preprocessor.process_dataset(
    csv_path='../01_data/splits/train.csv',
    max_samples=10  # Start with 10 samples for testing
)
```
